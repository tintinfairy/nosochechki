## Вариант 1 (сложный)

```python
from itertools import permutations
import re
```

```python
str = "носочечки"
```

```python
# True, если в строке есть последовательные гласные или согласные
# False, если нет
def has_consec_cons_or_vow(s):
    pattern_cons = re.compile(r'[нсчк]{2}')
    match_cons = pattern_cons.search(s)
    pattern_vow = re.compile(r'[оеи]{2}')
    match_vow = pattern_vow.search(s)
    return bool(match_cons) or bool(match_vow)
```

```python
result = set()
for permutation in permutations(str, 9):
    permutation_str = ''.join(permutation) # конвертируем tuple в строку
    if (not has_consec_cons_or_vow(permutation_str)): # отфильтровываем строки, в которых нет последовательных гласных/согласных
        result.add(permutation_str) # добавляем результаты в set (set не содержит повторяющихся элементов)
```

```python
count = 0
for res in sorted(result):
    count += 1
    print(count, res) # выводим все комбинации
```

```python
print('Количество комбинаций: ', len(result))
```

## Вариант 2

```python
from itertools import permutations
import re
```

```python
str = "носочечки"
```

```python
result = set() 
for permutation in permutations(str, 9):
    permutation_str_init = ''.join(permutation) # сохраняем слово до замены букв цифрами
    permutation_str = permutation_str_init # создаём копию строки, в которой заменим согласные 0, а гласные 1
    permutation_str = permutation_str.replace('н', '0')
    permutation_str = permutation_str.replace('с', '0')
    permutation_str = permutation_str.replace('ч', '0')
    permutation_str = permutation_str.replace('к', '0')
    permutation_str = permutation_str.replace('о', '1')
    permutation_str = permutation_str.replace('е', '1')
    permutation_str = permutation_str.replace('и', '1')
    if '11' not in permutation_str and '00' not in permutation_str:
        # добавляем результаты в set (set не содержит повторяющихся элементов)
        # новый элемент не добавляется, если уже существует в наборе
        result.add(permutation_str_init) 
```

```python
print('Количество комбинаций: ', len(result))
```
